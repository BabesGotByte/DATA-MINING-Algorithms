```
import pandas
import numpy
import matplotlib.pyplot as plt
from sklearn.cluster import Birch
```

```
datalist = pandas.read_csv('Dataset.txt',header = None,sep = "\s+|\t+|\s+\t+|\t+\s+",names=['a','b'])
datalist = datalist[::150]
```

NUMBER OF CLUSTERS = 10

```
birch = Birch(threshold=0.17,branching_factor=2,n_clusters=10)
```

```
birch.fit(datalist)
birch.predict(datalist)
```

```
plt.scatter(datalist.a,datalist.b,c=birch.predict(datalist))
```

NUMBER OF CLUSTERS = 5

```
birch = Birch(threshold=0.17,branching_factor=3,n_clusters=5)
```

```
birch.fit(datalist)
birch.predict(datalist)
```

```
plt.scatter(datalist.a, datalist.b, c=birch.predict(datalist))
```
