1. Reading a dataset into a dataframe

```
import numpy as np
import pandas as pd

df = pd.read_csv("weather.csv")
print(df)
df.head(10)
```

2. Dropping rows with missing(”NaN”) values

```
df.dropna("rows")
```

3. Dropping columns with missing(”NaN”) values

```
df.dropna("columns")
```

4. Filling the ”Nan” values with mean, median

```
df.fillna(df.mean())
df.fillna(df.median())
```

5. Split data set by row and column wise


```
#Split values by a character:

print(df.RainToday.str.split('e').str[0].tolist())
print(df.RainToday.str.split('e').str[1].tolist())
```
